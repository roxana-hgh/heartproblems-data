# Heart Problems

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets_heart.csv')
df['age'] = (df['age'] / 365).astype(int)
df = df.drop(['id','ap_hi','ap_lo','gluc','cardio'],axis = 1)
```

## Gender
#### It's shows that throughout life, men were about twice as likely as women to have a heart attack.

```python
mens = df[df['gender']== 1].shape[0]
womens = df[df['gender']== 2].shape[0]
plt.figure(figsize=(18,6))
plt.pie([mens,womens],labels=['Men','Women'],shadow=True,autopct='%1.1f%%',colors=['#1F77B4','#ed4e4e']);
```

## Ages

```python
ages = df['age']
m_less30 = ages[(df.age < 40) & (df.gender == 1) ].shape[0]
m_b30_40 = ages[(df['age'] < 40) & (df['age'] >= 30) & (df.gender == 1)].shape[0]
m_b40_50 = ages[(df['age'] < 50) & (df['age'] >= 40) & (df.gender == 1)].shape[0]
m_b50_60 = ages[(df['age'] < 60) & (df['age'] >= 50) & (df.gender == 1)].shape[0]
m_more60 = ages[(df['age'] >= 60) & (df.gender == 1)].shape[0]

w_less30 = ages[(df.age < 40) & (df.gender == 2) ].shape[0]
w_b30_40 = ages[(df['age'] < 40) & (df['age'] >= 30) & (df.gender == 2)].shape[0]
w_b40_50 = ages[(df['age'] < 50) & (df['age'] >= 40) & (df.gender == 2)].shape[0]
w_b50_60 = ages[(df['age'] < 60) & (df['age'] >= 50) & (df.gender == 2)].shape[0]
w_more60 = ages[(df['age'] >= 60) & (df.gender == 2)].shape[0]

mens_ages = [m_less30,m_b30_40,m_b40_50,m_b50_60,m_more60]
womens_ages = [w_less30,w_b30_40,w_b40_50,w_b50_60,w_more60]
grop = ["Less than 30" , "30 - 40", "40 - 50","50 - 60", "More than 60"]
```

```python
fig, ax = plt.subplots(figsize=(10,6))
width = 0.35
x = np.arange(len(grop))
mens = ax.bar(x - width/2, mens_ages, width, label='Men');
womens = ax.bar(x + width/2, womens_ages, width, label='Women',color = '#ed4e4e');

ax.set_ylabel('people')
ax.set_title('Ages by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(grop)
ax.legend();

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(mens)
autolabel(womens)
```

## Smoking

```python
smoke = df['smoke'].value_counts()
no = smoke[0]
yes = smoke[1]
plt.figure(figsize=(18,6))
plt.pie([no,yes],labels=['Did not smoke','smoked'],explode=(0,0.05),shadow=True ,autopct='%1.1f%%',colors=['#9ACD32','#655555']);
```
